## Cuestión 6:  ¿Cuál de estas funciones de activación no debería usarse en una capa oculta (hidden layer)?

**a)** `sigmoid`

**b)** `tanh`

**c)** `relu`

**d)** `linear`


**d) linear**

Justificación:

1. Funciones de activación en capas ocultas:
   * Las funciones de activación como sigmoid, tanh y relu introducen no linealidad en el modelo, lo que permite a la red aprender relaciones complejas y no lineales en los datos.
   * Estas funciones son esenciales para que las redes neuronales profundas puedan aproximar funciones complejas.
2. Por qué no usar linear:
   * Una función de activación lineal no introduce no linealidad, lo que significa que las capas ocultas no agregarían valor adicional al modelo.
   * Si todas las capas ocultas usaran una función lineal, la red completa se comportaría como una sola capa lineal, independientemente de la profundidad de la red.

**Conclusión:**
  * ***linear*** no debería usarse en capas ocultas porque elimina la capacidad de la red para aprender patrones no lineales, lo que es fundamental en problemas complejos.

## Cuestión 8:  Supongamos que queremos entrenar una red para un problema de clasificación de imágenes con las siguientes clases: {'perro','gato','persona'}. ¿Cuántas neuronas y que función de activación debería tener la capa de salida? ¿Qué función de pérdida (loss function) debería usarse?


**Capa de salida: Número de neuronas y función de activación**
1. Número de neuronas:
   * Dado que el problema tiene 3 clases ({'perro', 'gato', 'persona'}), la capa de salida debe tener 3 neuronas, una para cada clase.
2. Función de activación:
   * La función de activación más adecuada para un problema de clasificación multiclase es Softmax.
   * Softmax convierte las salidas de las neuronas en probabilidades (valores entre 0 y 1) que suman 1, lo que es ideal para problemas de clasificación multiclase.

**Función de pérdida (loss function):**
  * La función de pérdida adecuada para un problema de clasificación multiclase es la Categorical Cross-Entropy.
  * Si las etiquetas están codificadas como one-hot vectors (por ejemplo, [1, 0, 0] para "perro"), se usa directamente la Categorical Cross-Entropy.
  * Si las etiquetas están codificadas como enteros (por ejemplo, 0 para "perro", 1 para "gato"), se usa la Sparse Categorical Cross-Entropy, que es una variante optimizada para este caso.
